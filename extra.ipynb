```python
#Feature Encoding:Convert the categorical data into numerical data
import pandas as pd
info = {
    "Marks":['Good','Average','Bad','Average','Good','Bad','Average','Good','Bad','Average'],
    "Grade":['A','B','C','B','A','C','B','A','C','B']
}
df = pd.DataFrame(info)
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
marks_encoded = le.fit_transform(df['Marks'])
print(marks_encoded)
```

```python
df['Marks'] = marks_encoded
print(df)
```

```python
grade_encoded = le.fit_transform(df['Grade'])
print(grade_encoded)
```

```python
df['Grade'] = grade_encoded
print(df)
```
